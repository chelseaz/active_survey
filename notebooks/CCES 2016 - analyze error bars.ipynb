```r
n_reps <- 50
all_results <- bind_rows(lapply(seq(1, n_reps), function(rep) {
    cur_results <- read_results(paste0('cces16_', rep, '_bpmf'))
    IRdisplay::display(get_sample_sizes(cur_results))
    
    cur_results %>% 
        filter(columns == 'all') %>%
        select(strategy, qnum, mse, mae, pws, bias)
}))
```

```r
all_results %>% group_by(strategy, qnum) %>% summarize(n = n())
```

```r
all_results_for_plot <- all_results %>% 
    mutate(qnum = as.factor(qnum))  # otherwise R won't create separate boxplots
options(repr.plot.width=16, repr.plot.height=16)
ggplot(all_results_for_plot) + 
    geom_violin(aes_string(x="qnum", y="mse", fill="strategy")) +
    xlab("Number of questions")
```

```r
# how well do mean and error bars line up with this?
all_results_long <- all_results %>% 
    gather(metric, value, -strategy, -qnum)

all_results_agg <- all_results_long %>%
    group_by(strategy, qnum, metric) %>%
    summarize(n = n(), mean = mean(value), sd = sd(value)) %>%
    mutate(interval_min = mean-2*sd, interval_max = mean+2*sd)
```

```r
interval_coverage <- all_results_long %>% 
    inner_join(all_results_agg, by = c("strategy", "qnum", "metric")) %>%
    mutate(in_interval = (value >= interval_min & value <= interval_max)) %>%
    group_by(strategy, qnum, metric) %>%
    summarize(coverage = mean(in_interval))

options(repr.plot.width=6, repr.plot.height=4)

ggplot(interval_coverage) + 
    geom_line(aes_string(x="qnum", y="coverage")) +
    facet_grid(metric ~ strategy) +
    xlab("Number of questions") +
    ylab("Coverage of confidence interval")
```

```r
options(repr.plot.width=6, repr.plot.height=4)

ggplot(interval_coverage) + 
    geom_histogram(aes_string(x="coverage"), binwidth=0.01) +
    facet_grid(metric ~ strategy) +
    xlab("Coverage of confidence interval") + 
    ylab("Question count")
```
